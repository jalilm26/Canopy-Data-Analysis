```python
print(df3.columns)
```

```python
#display all available genres in the french language
french_genres = df3[df3['Language'] == 'French']['Genres'].unique()
print("Genres available in French language:", french_genres)
```

```python
#total number of genres 
total_genres_french = len(french_genres)
print("Total count of genres in French language:", total_genres_french)
```

```python
#top 20 french movies sorted by rotten tomato score
french_movies = df3[df3['Language'] == 'French']
movie_genre = df3[df3['Genres'] == 'Comedy']
sorted_movies = french_movies.sort_values(by='Rotten Tomatoes', ascending=False)
selected_columns = ['Title', 'Language', 'Rotten Tomatoes', 'Genres', 'Age']
top_20 = sorted_movies.head(20)
print(top_20[selected_columns])

```

```python
#find out how many french movies there are per genre
french_movies1 = df3[df3['Language'] == 'French']
movies_per_genre = french_movies1['Genres'].value_counts()
print(movies_per_genre)
```
